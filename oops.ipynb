```python
class books:
    def __init__(self):
        self.book=["The Alchemist","Rich Dad Poor Dad","Atomic Habits"]
        self.author=["Paulo Coelho","Robert T. Kiyosaki"," James Clear"]
    def add_book(self,books):
        self.book.append(books)
    def add_author(self,authors):
        self.author.append(authors)



class User:
    def __init__(self):
        self.user_name=[]
        self.user_passw=[]
        self.cart=[]
    def sing(self,name,passw):
        self.user_name.append(name)
        self.user_passw.append(passw)
    def login(self,name,passw):
        for i in range(len(self.user_name)):
            if self.user_name[i] == name and self.user_passw[i] == passw:
                print("You can enter")
                enter()
                return
        return print("You can't enter")
    def add_cart(self,book):
        self.cart.append(book)



user_ = User()
books_ = books()



def exit():
    print("exit")



def add_book():
    n=int(input("enter how many book you need to add"))
    for i in range(n):
        b=input("enter book name")
        a=input("enter author name")
        books_.add_book(b)
        books_.add_author(a)



def admin():
    c5=int(input("1.show book\n2.add book\n3.exit"))
    if c5==1:
        for i in range(len(books_.book)):
            print(f"{books_.book[i]} - {books_.author[i]}")
    elif c5==2:
        add_book()
    else:
        exit()



def pick_book():
    n=int(input("enter how many book you need to add in cart\n"))
    for j in range(n):
        book_name=input("enter a book name")
        for i in range(len(books_.book)):
            if books_.book[i]==book_name:
                user_.add_cart(books_.book[i])
    c4=int(input("1.view cart \n2.exit\n"))
    if c4==1:
        for i in range(len(user_.cart)):
            print(user_.cart[i])
    else:
        exit()



def enter():
    c3=int(input("1.show books \n2.exit\n"))
    if c3==1:
        for i in range(len(books_.book)):
            print(f"{books_.book[i]} - {books_.author[i]}")
        pick_book()
    else:
        exit()



def user():
    c2=int(input("1.sing in \n2.login\n"))
    if c2==1:
        n=int(input("enter how many you need to enter"))
        for i in range(n):
            name=input("enter name")
            passw=int(input("enter password"))
            user_.sing(name,passw)
    name=input("enter name")
    passw=int(input("enter password"))
    user_.login(name,passw)



c1=int(input("1.User \n2.admin \n3.exit\n"))
if c1==1:
    user()

elif c1==2:
    admin()

else:
    exit()
```
